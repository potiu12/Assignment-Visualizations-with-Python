```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
covid_df = pd.read_csv(url, index_col=0)
covid_df
```

```python
#%% viz 1
'''
Create a visualization that shows all of the counties in Utah as a time series,
similar to the one shown in slide 22 during the lecture. The graphic should
-Show cases over time
-Have all counties plotted in a background color (something like grey)
-Have a single county plotted in a contrasting color (something not grey)
-Have well formatted dates as the X axis
'''
```

```python
utah_data = covid_df[covid_df['Province_State'] == 'Utah']
utah_data
```

```python
#%% viz 1

# Define the function to extract date columns
def extract_date_columns(dataframe):
    """
    Extract columns that represent dates by checking if they can be parsed as valid dates.
    """
    date_columns = []
    for col in dataframe.columns:
        try:
            pd.to_datetime(col, format='%m/%d/%y', errors='raise')
            date_columns.append(col)
        except ValueError:
            continue
    return date_columns

# Use the function
date_columns = extract_date_columns(utah_data)

# Plot setup
plt.figure(figsize=(12, 6))

# Plot all counties in Utah in grey
for county in utah_data.index:
    plt.plot(date_columns, utah_data.loc[county, date_columns], color='grey', alpha=0.6, label='_nolegend_')

# Highlight one county (e.g., Salt Lake County)
highlight_county = 'Salt Lake'
highlight_data = utah_data[utah_data['Admin2'] == highlight_county]
plt.plot(date_columns, highlight_data.iloc[0][date_columns], color='red', linewidth=2, label=highlight_county)

# Add labels, title, and legend
plt.title('COVID-19 Cases Over Time in Utah Counties', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Cumulative Cases', fontsize=12)
plt.xticks(ticks=range(0, len(date_columns), 30), labels=date_columns[::30], rotation=45)

plt.legend(loc='upper left', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Rename columns to ensure consistency
covid_df.rename(columns={
    'Province_State': 'State',
    'Admin2': 'County'
}, inplace=True)

# Identify date columns for calculating daily cases
date_columns = [col for col in covid_df.columns if '/' in col]
date_columns.sort()

# Calculate daily cases if not already calculated
if 'Daily Cases' not in covid_df.columns:
    covid_df['Daily Cases'] = covid_df[date_columns].diff(axis=1).iloc[:, -1].fillna(0)

# Sort states by total cases
state_order = covid_df.groupby('State')['Daily Cases'].sum().sort_values(ascending=False).index

# Ensure the necessary columns exist
if 'State' in covid_df.columns and 'Daily Cases' in covid_df.columns:
    # Create grouped boxplot for all states
    sns.set_theme(style="ticks", palette="pastel")
    plt.figure(figsize=(25, 10))  # Increase figure size to accommodate all states

    sns.boxplot(
        x='State',
        y='Daily Cases',
        data=covid_df,
        order=state_order,
        palette="muted"
    )

    # Adjust y-axis to focus on the majority of the data (e.g., exclude extreme outliers)
    y_limit = covid_df['Daily Cases'].quantile(0.95)  # Get the 95th percentile value
   # Limit the y-axis range
    plt.ylim(0, 200000)


    # Adjust labels and title
    plt.title('Grouped Box Plot of Daily Cases by State (95th Percentile)', fontsize=18)
    plt.xlabel('State', fontsize=14)
    plt.ylabel('Daily Cases', fontsize=14)
    plt.xticks(rotation=90, fontsize=10)  # Rotate x-axis labels for better readability

    # Finalize and show the plot
    plt.tight_layout()
    plt.show()
else:
    print("Error: The required columns 'State' and 'Daily Cases' are not in the DataFrame.")
```
